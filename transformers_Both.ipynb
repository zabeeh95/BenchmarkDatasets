```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

eng_sentences = ["<start> " + s + " <end>" for s in eng_sentences]
urdu_sentences = ["<start> " + s + " <end>" for s in urdu_sentences]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(eng_sentences)
tokenizer.fit_on_texts(urdu_sentences)

eng_seq = tokenizer.texts_to_sequences(eng_sentences)
urdu_seq = tokenizer.texts_to_sequences(urdu_sentences)

max_len_eng = max(len(seq) for seq in eng_seq)
max_len_urdu = max(len(seq) for seq in urdu_seq)

eng_seq = pad_sequences(eng_seq, maxlen=max_len_eng, padding='post')
urdu_seq = pad_sequences(urdu_seq, maxlen=max_len_urdu, padding='post')

encoder_input = eng_seq
decoder_input = urdu_seq[:, :-1]
decoder_output = urdu_seq[:, 1:]

vocab_size = len(tokenizer.word_index) + 1
print(f"vocab size is  : {vocab_size}")

print(f"total sequences are {len(eng_seq) + len(urdu_seq)} -- eng seq : {max_len_eng} -- urdu seq :{max_len_urdu}  ")
print(f"encoder_input shape: {encoder_input.shape}")  # shape: [num_sentences, max_len]
print(f"decoder_input shape: {decoder_input.shape}")
print(f"decoder_output shape: {decoder_output.shape}")
```

```python
# padding_mask = tf.cast(y != 0, tf.float32)  # shape: [num_sentences, max_len-1]

def create_causal_mask(seq_len):
    return tf.linalg.band_part(tf.ones((seq_len, seq_len)), -1, 0)


def positional_encoding(seq_len, d_model):
    pos = np.arange(seq_len)[:, np.newaxis]
    i = np.arange(d_model)[np.newaxis, :]

    angle_rates = 1 / np.power(10000, (2 * (i // 2)) / d_model)
    angle_rads = pos * angle_rates

    pos_encoding = np.zeros((seq_len, d_model))
    pos_encoding[:, 0::2] = np.sin(angle_rads[:, 0::2])
    pos_encoding[:, 1::2] = np.cos(angle_rads[:, 1::2])

    return tf.cast(pos_encoding, tf.float32)


class EncoderBlock(tf.keras.layers.Layer):
    def __init__(self, d_model, num_heads):
        super().__init__()

        self.mha = tf.keras.layers.MultiHeadAttention(num_heads=num_heads,
                                                      key_dim=d_model // num_heads)

        self.ffn = tf.keras.Sequential([
            tf.keras.layers.Dense(2 * d_model, activation='relu'),
            tf.keras.layers.Dense(d_model)
        ])

        self.norm1 = tf.keras.layers.LayerNormalization(epsilon=1e-6)
        self.norm2 = tf.keras.layers.LayerNormalization(epsilon=1e-6)

    def call(self, x):
        attn = self.mha(x, x)
        x = self.norm1(x + attn)

        ffn_out = self.ffn(x)
        return self.norm2(x + ffn_out)


class DecoderBlock(tf.keras.layers.Layer):
    def __init__(self, d_model, num_heads):
        super().__init__()

        # Masked self-attention
        self.mha1 = tf.keras.layers.MultiHeadAttention(num_heads=num_heads,
                                                       key_dim=d_model // num_heads)

        # Cross attention (encoder → decoder)
        self.mha2 = tf.keras.layers.MultiHeadAttention(num_heads=num_heads,
                                                       key_dim=d_model // num_heads)

        self.ffn = tf.keras.Sequential([
            tf.keras.layers.Dense(2 * d_model, activation='relu'),
            tf.keras.layers.Dense(d_model)
        ])

        self.norm1 = tf.keras.layers.LayerNormalization(epsilon=1e-6)
        self.norm2 = tf.keras.layers.LayerNormalization(epsilon=1e-6)
        self.norm3 = tf.keras.layers.LayerNormalization(epsilon=1e-6)

    def call(self, x, enc_output):
        seq_len = tf.shape(x)[1]
        mask = create_causal_mask(seq_len)

        # Masked self-attention
        # attn1 = self.mha1(x, x, attention_mask=mask)
        attn1 = self.mha1(query=x, key=x, value=x, attention_mask=mask)
        x = self.norm1(x + attn1)

        # Cross attention (IMPORTANT)
        # attn2 = self.mha2(x, enc_output)
        attn2 = self.mha2(query=x, key=enc_output, value=enc_output)
        x = self.norm2(x + attn2)

        # FFN
        ffn_out = self.ffn(x)
        return self.norm3(x + ffn_out)


class Transformer(tf.keras.Model):
    def __init__(self, vocab_size, max_len_enc, max_len_dec, d_model=128, num_heads=4):
        super().__init__()

        # Embeddings
        self.enc_embedding = tf.keras.layers.Embedding(vocab_size, d_model)
        self.pos_enc = positional_encoding(max_len_enc, d_model)
        self.encoder = EncoderBlock(d_model, num_heads)

        self.dec_embedding = tf.keras.layers.Embedding(vocab_size, d_model)
        self.pos_dec = positional_encoding(max_len_dec, d_model)
        self.decoder = DecoderBlock(d_model, num_heads)

        self.final_layer = tf.keras.layers.Dense(vocab_size)

    def call(self, inputs):
        enc_inp, dec_inp = inputs

        # Encoder  [200 8]
        enc = self.enc_embedding(enc_inp)  # [200 8 128]
        enc += self.pos_enc[:tf.shape(enc)[1], :]  # [8 128] + [200 8 128] - > [200 8 128]
        enc_output = self.encoder(enc)  # [200 8 128]

        # Decoder
        dec = self.dec_embedding(dec_inp)
        dec += self.pos_dec[:tf.shape(dec)[1], :]
        dec_output = self.decoder(dec, enc_output)

        return self.final_layer(dec_output)  # [200 8 573]

    # def call(self, inputs):
    #     enc_inp, dec_inp = inputs
    #
    #     # Encoder input shape
    #     tf.print("Encoder input shape:", tf.shape(enc_inp))  # (batch, seq_len_enc)
    #
    #     # Decoder input shape
    #     tf.print("Decoder input shape:", tf.shape(dec_inp))  # (batch, seq_len_dec)
    #
    #     # Encoder embeddings
    #     enc = self.enc_embedding(enc_inp)
    #     tf.print("Encoder token embedding shape:", tf.shape(enc))  # (batch, seq_len_enc, d_model)
    #
    #     # Encoder positional encoding
    #     pos = self.pos_enc[:tf.shape(enc)[1], :]
    #     tf.print("Encoder positional encoding shape:", tf.shape(pos))  # (seq_len_enc, d_model)
    #
    #     # Combine embeddings + pos
    #     enc_pos_combine = enc + pos
    #     tf.print("Encoder combined embedding shape:", tf.shape(enc_pos_combine))  # (batch, seq_len_enc, d_model)
    #
    #     enc_output = self.encoder(enc_pos_combine)
    #     tf.print("Encoder output shape:", tf.shape(enc_output))  # (batch, seq_len_enc, d_model)
    #
    #     # Decoder embeddings
    #     dec = self.dec_embedding(dec_inp)
    #     tf.print("Decoder token embedding shape:", tf.shape(dec))  # (batch, seq_len_dec, d_model)
    #
    #     dec_pos = self.pos_dec[:tf.shape(dec)[1], :]
    #     tf.print("Decoder positional encoding shape:", tf.shape(dec_pos))  # (seq_len_dec, d_model)
    #
    #     dec += dec_pos
    #     tf.print("Decoder combined embedding shape:", tf.shape(dec))  # (batch, seq_len_dec, d_model)
    #
    #     dec_output = self.decoder(dec, enc_output)
    #     tf.print("Decoder output shape:", tf.shape(dec_output))  # (batch, seq_len_dec, d_model)
    #
    #     final = self.final_layer(dec_output)
    #     tf.print("Final output shape:", tf.shape(final))  # (batch, seq_len_dec, vocab_size)
    #
    #     return final


model = Transformer(vocab_size,
                    max_len_eng,
                    max_len_urdu - 1)  # urdu length -1 - > because decoder_input is shorter
```

```python
index_word = tokenizer.index_word
word_index = tokenizer.word_index


def translate_sentence(model, sentence, max_len_urdu):
    # Add start/end
    sentence = "<start> " + sentence + " <end>"

    # Convert to sequence
    seq = tokenizer.texts_to_sequences([sentence])
    seq = pad_sequences(seq, maxlen=max_len_eng, padding='post')

    encoder_input = tf.convert_to_tensor(seq)

    # Start token for decoder
    start_token = word_index["start"] if "start" in word_index else word_index["<start>"]
    end_token = word_index["end"] if "end" in word_index else word_index["<end>"]

    decoder_input = [start_token]

    output_sentence = []

    for i in range(max_len_urdu):
        dec_input = pad_sequences([decoder_input], maxlen=max_len_urdu - 1, padding='post')

        predictions = model([encoder_input, dec_input])

        # Take last timestep
        predicted_id = tf.argmax(predictions[0, i]).numpy()

        # Stop if <end>
        if predicted_id == end_token:
            break

        decoder_input.append(predicted_id)

        if predicted_id in index_word:
            output_sentence.append(index_word[predicted_id])

    return " ".join(output_sentence)


print(translate_sentence(model, "i love you", max_len_urdu))
print(translate_sentence(model, "where are you?", max_len_urdu))
print(translate_sentence(model, "i am hungry", max_len_urdu))
```
